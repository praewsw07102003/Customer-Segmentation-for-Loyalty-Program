```python
import pandas as pd
try:
    df = pd.read_csv('/content/drive/MyDrive/Customer Segmentation for Loyalty Program/data.csv', encoding='utf-8')
except UnicodeDecodeError:
    print("UTF-8 encoding failed, trying 'latin1'")
    df = pd.read_csv('/content/drive/MyDrive/Customer Segmentation for Loyalty Program/data.csv', encoding='latin1')
df
```

```python
print(df.info())

print(df.describe())

print("Missing values ต่อคอลัมน์:")
print(df.isnull().sum())

print("จำนวนค่าที่ไม่ซ้ำ:")
print(df.nunique())
```

```python
df = df.dropna(subset=['CustomerID'])
df['CustomerID'] = df['CustomerID'].astype(int).astype(str)
df = df[~df['InvoiceNo'].astype(str).str.startswith('C')]
df = df[df['Quantity'] > 0]
df = df[df['UnitPrice'] > 0]
df['InvoiceDate'] = pd.to_datetime(df['InvoiceDate'], errors='coerce')
df['TotalPrice'] = df['Quantity'] * df['UnitPrice']
```

```python
df
```

```python
df['StockCode'] = df['StockCode'].str.replace(r'[^\d]', '', regex=True)
df = df[df['StockCode'].str.strip() != '']
df
```

```python
print("ข้อมูลหลัง Clean แล้ว:")
print(df.info())
print(df.head())
```

```python
df = df[~df['Country'].isin(['Unspecified', 'European Community'])]

df = df.drop_duplicates()
```

```python
negative_quantity = df[df['Quantity'] <= 0]
print(f"จำนวนแถวที่ Quantity <= 0: {len(negative_quantity)}")
print(negative_quantity.head())

duplicate_invoice = df[df.duplicated(subset=['InvoiceNo', 'StockCode', 'Quantity', 'InvoiceDate'], keep=False)]
print(f"จำนวนแถวที่ InvoiceNo ซ้ำพร้อมข้อมูลซ้ำเหมือนกัน: {len(duplicate_invoice)}")
print(duplicate_invoice.head())

invoice_counts = df['InvoiceNo'].value_counts()
duplicate_invoice_nos = invoice_counts[invoice_counts > 1]
print(f"จำนวน InvoiceNo ที่ซ้ำมากกว่า 1 ครั้ง: {len(duplicate_invoice_nos)}")
print(duplicate_invoice_nos.head())

empty_description = df[df['Description'].isna() | (df['Description'].str.strip() == "")]
print(f"จำนวนแถวที่ Description ว่าง: {len(empty_description)}")
print(empty_description.head())

import re

def is_stockcode_numeric(s):
    return bool(re.fullmatch(r'\d+', s))

non_numeric_stockcode = df[~df['StockCode'].apply(is_stockcode_numeric)]
print(f"จำนวนแถวที่ StockCode ไม่ใช่ตัวเลขล้วน: {len(non_numeric_stockcode)}")
print(non_numeric_stockcode.head())

empty_country = df[df['Country'].isna() | (df['Country'].str.strip() == "")]
print(f"จำนวนแถวที่ Country ว่าง: {len(empty_country)}")
print(empty_country.head())

unique_countries = df['Country'].unique()
print(f"รายชื่อประเทศทั้งหมด ({len(unique_countries)}):")
print(unique_countries)
```

```python
import pandas as pd
snapshot_date = df['InvoiceDate'].max() + pd.Timedelta(days=1)

rfm = df.groupby('CustomerID').agg({
    'InvoiceDate': lambda x: (snapshot_date - x.max()).days,  # Recency
    'InvoiceNo': 'nunique',                                    # Frequency
    'TotalPrice': 'sum'                                        # Monetary
})

rfm.columns = ['Recency', 'Frequency', 'Monetary']

print("RFM Table ตัวอย่าง:")
rfm.head()
```

```python
from google.colab import files
rfm.to_excel("rfm_segmented.xlsx")
files.download("rfm_segmented.xlsx")
```

```python
from sklearn.preprocessing import StandardScaler
from sklearn.cluster import KMeans

X = rfm[['Recency', 'Frequency', 'Monetary']]
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

kmeans = KMeans(n_clusters=4, random_state=42)
rfm['Cluster'] = kmeans.fit_predict(X_scaled)
```

```python
def rfm_segment(row):
    if row['Recency'] <= 30 and row['Frequency'] >= 5 and row['Monetary'] >= 500:
        return 'VIP'
    elif row['Recency'] > 90 and row['Frequency'] <= 2:
        return 'At Risk'
    elif row['Frequency'] == 1 and row['Monetary'] < 200:
        return 'New'
    else:
        return 'Regular'

rfm['Group'] = rfm.apply(rfm_segment, axis=1)
```

```python
cluster_order = rfm.groupby('Cluster')['Recency'].mean().sort_values().index.tolist()
cluster_names = {cluster_order[0]: 'VIP',
                 cluster_order[1]: 'Loyal',
                 cluster_order[2]: 'Regular',
                 cluster_order[3]: 'At Risk'}

rfm['Group'] = rfm['Cluster'].map(cluster_names)
```
